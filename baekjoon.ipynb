```python
##problem 9037

##같은 수인지 체크
def check(arr, n):
    for i in range(n):
        if arr[i] % 2:
            arr[i] += 1
    return len(set(arr)) == 1

##분배
def teacher(arr, n):
    tmp_list = [0] * n
    for i in range(n):
        if arr[i] % 2:
            tmp_list[(i+1) % n] = (arr[i]+1)//2
        else:
            tmp_list[(i+1) % n] = arr[i]//2
        arr[i] //= 2
    for i in range(n):
        arr[i] += tmp_list[i]
    return arr
    
def same_amount():
    n, candy = int(input()), list(map(int, input().split()))
    result = 0
    while not check(candy, n):
        candy = teacher(candy, n)
        result += 1
    return result

t = int(input())
for _ in range(t):
    print(same_amount())
```

```python
##problem1074
# 1<=N<=15, 0<= r,c <= 2^N-1
N,r,c = map(int, input().split())

def z(sz, x, y):
    if sz == 1:
        return 0
    sz //= 2
    for i in range(2):
        for j in range(2):
            if x < sz * (i+1) and y < sz * (j+1):
                return (i*2+j) * sz * sz + z(sz, x-sz*i, y-sz*j)
            
print(z(2**N,r,c))
```

```python
##problem 2484
# 같은 눈 4개: 50000+같은 눈 * 5000
# 같은 눈 3개: 10000+같은 눈 * 1000
# 같은 눈 2개씩 2개 : 2000+ 2개가 나온 눈*500 + 2개가 나온 눈 500
# 같은 눈이 2개만 나오는 경우: 1000 + 같은 눈*100
# 모두 다른 경우: 가장 큰 눈* 100


def calculate(input_v):
    result = 0
    cnt = 5 - len(set(input_v))
    if cnt == 4:
        result = input_v[0]*5000+50000
    
    elif cnt == 3:
        if input_v[1] != input_v[2]: ## 겹치는 눈이 3개인 경우
            result = sum(set(input_v))*500+2000
        else: ## 겹치는 눈이 2개씩 2쌍인 경우
            result = input_v[1]*1000+10000
    elif cnt == 2:
        if input_v[0] == input_v[1]:
            result =input_v[0]*100+1000
        elif input_v[1] == input_v[2]:
            result =input_v[1]*100+1000
        else:
            result = input_v[2]*100+1000
    else:
        result = input_v[3] * 100
    return result
n = int(input())
sums = []
for _ in range(n):
    dice = sorted(list(map(int, input().split())))
    sums.append(calculate(dice))
print(max(sums))
```

```python
#17413
# 1< len(S) < 100000 
#경우의 수를 나눠서 보자 문자열의 경우 특히 복잡함
ck, tmp, ans, S = False,'','',input()

for i in S:
    if i == ' ':
        ans += tmp[::-1] + ' '
        tmp = ''
    elif i == '<':
        ans += tmp[::-1] + '<'
        tmp = ''
        ck = True
    elif i == '>':
        ans += '>'
        ck = False
    else:
        if ck:
            ans += i
        else:
            tmp += i
            
print(ans+tmp[::-1])
```
