```python
##sort2
##알고리즘 수업의 코드를 가져옴

def merge(left, right):
    merged = list()
    left_point, right_point = 0, 0
    
    # case1 - left/right 둘다 있을때
    while len(left) > left_point and len(right) > right_point:
        if left[left_point] > right[right_point]:
            merged.append(right[right_point])
            right_point += 1
        else:
            merged.append(left[left_point])
            left_point += 1

    # case2 - left 데이터가 없을 때
    while len(left) > left_point:
        merged.append(left[left_point])
        left_point += 1
        
    # case3 - right 데이터가 없을 때
    while len(right) > right_point:
        merged.append(right[right_point])
        right_point += 1
    
    return merged


def mergesplit(data):
    if len(data) <= 1:
        return data
    medium = int(len(data) / 2)
    left = mergesplit(data[:medium])
    right = mergesplit(data[medium:])
    return merge(left, right)

need_sort = list()
        
n = int(input())
for _ in range(n):
    insert_number = int(input())
    need_sort.append(insert_number)
array_list = mergesplit(need_sort)
    
for n in array_list:
    print(n)
```

```python
##sort2
##quick sort는 최악의 경우 O(N^2)의 시간복잡도를 가진다
##따라서 O(NlogN)의 시간복잡도를 가지는 병합정렬 힙정렬을 사용하자

def merge_sort(array):
    length = len(array)
    if length == 1:
        return array
    mid = length//2
    left = array[:mid]
    right = array[mid:]
    left_list = merge_sort(left)
    right_list = merge_sort(right)
    left_idx = 0
    right_idx = 0
    
    temp = list()
    while left_idx < len(left_list) and right_idx < len(right_list):
        if left_list[left_idx] < right_list[right_idx]:
            temp.append(left_list[left_idx])
            left_idx += 1
        else:
            temp.append(right_list[right_idx])
            right_idx += 1
        if left_idx == len(left_list):
            temp.extend(right_list[right_idx:])
            break
        elif right_idx == len(right_list):
            temp.extend(left_list[left_idx:])
            break
    return temp
        

need_sort = list()
        
n = int(input())
for _ in range(n):
    insert_number = int(input())
    need_sort.append(insert_number)
array_list = merge_sort(need_sort)
    
for n in array_list:
    print(n)
    
```

```python
def z(N, r, c):
    if N == 1:
        if (r, c) == (0, 0):
            return 0
        elif (r, c) == (0, 1):
            return 1
        elif (r, c) == (1, 0):
            return 2
        elif (r, c) == (1, 1):
            return 3
    sqrt = 2 ** (N-1)
    if sqrt > r:
        if sqrt > c:
            return z(N-1, r, c)
        elif sqrt <= c:
            return sqrt ** 2 + z(N-1, r, c-sqrt)
    
    elif sqrt <= r:
        if sqrt > c:
            return sqrt ** 2 * 2 + z(N-1, r-sqrt, c)
        elif sqrt <= c:
            return sqrt ** 2 * 3 + z(N-1, r-sqrt, c-sqrt)
        

N, r, c = list(map(int, input().split(' ')))

print(z(N, r, c))
```

```python
def union(node1, node2):
    if rank[node1] > rank[node2]:
        parent[node2] = node1
    elif rank[node1] < rank[node2]:
        parent[node1] = node2
    else:
        parent[node2] = node1
        rank[node1]+=1
    

def find(node):
    if parent[node] != node:
        parent[node] = find(parent[node])
    return parent[node]

def make_set(node):
    parent[node] = node
    rank[node] = 0

def count_network(nodes):
    cnt = 0
    node1, node2 = nodes
    
    if node1 not in parent:
        make_set(node1)
    if node2 not in parent:
        make_set(node2)
    root_node1 = find(node1)
    root_node2 = find(node2)
    if root_node1 != root_node2:
        union(root_node1, root_node2)
    
    for value in parent:
        if find(root_node1) == find(value):
            cnt += 1
    
    return cnt
    
test_case = int(input())

for _ in range(test_case):
    nets = int(input())
    rank = dict()
    parent = dict()
    for _ in range(nets):
        print(count_network(input().split(' ')))
    
  

 
```
